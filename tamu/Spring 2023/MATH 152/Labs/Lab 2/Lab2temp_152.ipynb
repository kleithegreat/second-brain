### Question 1

```python
#start code here
x, y = symbols("x y")

equation = Eq(x, -1 * y**2 + 5*y - 3)
plot_implicit(equation, (x, -10, 10), (y, -2, 5))
```

#### 1b

```python
#start code here
print(solve(equation.subs(x, 0), y))
```

```python
volume = integrate(pi * (-1 * y**2 + 5*y - 3)**2, (y, solve(equation.subs(x, 0), y)[0], solve(equation.subs(x, 0), y)[1]))
print(N(volume))
```
